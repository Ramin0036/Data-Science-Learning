```python
import nltk
```

```python
text = ''' The teams growth this season had a huge impact, helping them break a historic record, and this record will likely impact the future growth of the club '''
```

```python
from nltk import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
```

```python
sen_tokens = sent_tokenize(text.lower())
```

```python
tokens = word_tokenize(text)
```

```python
sent_tokenize(text)
```

```python
stop_words = stopwords.words('english')
```

```python
filterd_sent = []
for token in tokens:
    if token not in stop_words:
        filterd_sent.append(token)
```
