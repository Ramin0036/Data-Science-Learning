```python
Sport = ''' The teams growth this season had a huge impact, helping them break a historic record, and this record will likely impact the future growth of the club '''
```

```python
Economy = ''' The country’s economic growth reached a record level this year, and this record growth is expected to have a strong impact on investment, employment, and long‑term economic growth '''
```

```python
News = ''' The report highlighted the rapid growth of renewable energy, reaching a record level this year, and this record growth could have a major global impact on climate policy and future energy growth '''
```

```python
document = [Sport , Economy , News]
label = ['sport' , 'economy' , 'news']
```

```python
from sklearn.feature_extraction.text import CountVectorizer , TfidfVectorizer
```

```python
vectorizer = CountVectorizer()
vec_data = vectorizer.fit_transform(document)
```

```python
vec_data
```

```python
vec_data_array = vec_data.toarray()
```

```python
vec_data_array
```

```python
features = vectorizer.get_feature_names_out()
```

```python
features
```

```python
import pandas as pd
```

```python
whole_data = pd.DataFrame(vec_data_array , columns=features)
```

```python
whole_data.shape
```

```python
whole_data
```

```python
Tf_vectorizer = TfidfVectorizer()
data_Tf = Tf_vectorizer.fit_transform(document)
```

```python
data_Tf
```

```python
data_Tf_array = data_Tf.toarray()
```

```python
data_Tf_array
```

```python
features_Tf = Tf_vectorizer.get_feature_names_out()
```

```python
features_Tf
```

```python
whole_data_Tf = pd.DataFrame(data_Tf_array , columns=features_Tf)
```

```python
whole_data_Tf
```
