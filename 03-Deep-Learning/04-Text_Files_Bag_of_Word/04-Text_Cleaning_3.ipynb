```python
from nltk import word_tokenize , sent_tokenize
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
Sport = ''' The teams growth this season had a huge impact, helping them break a historic record, and this record will likely impact the future growth of the club '''
Economy = ''' The countrys economic growth reached a record level this year, and this record growth is expected to have a strong impact on investment, employment, and long‑term economic growth '''
News = ''' The report highlighted the rapid growth of renewable energy, reaching a record level this year, and this record growth could have a major global impact on climate policy and future energy growth '''
```

```python
sport_token = word_tokenize(Sport)
economy_token = word_tokenize(Economy)
news_token = word_tokenize(News)
```

```python
stemmer = PorterStemmer()
```

```python
punctuations = list(string.punctuation)
```

```python
stop_words = stopwords.words('english')
```

```python
filterd_sent_soprt = []
for token in sport_token:
    if (token not in stop_words) and (token not in punctuations):
        word = stemmer.stem(token)
        filterd_sent_soprt.append(word)
```
