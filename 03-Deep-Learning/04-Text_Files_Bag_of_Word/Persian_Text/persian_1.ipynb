```python
text = 'سلام، من رامین الله وردی زاده اهل ارومیه؛ هستم؟'
```

```python
import hazm
import string
```

```python
normalizer = hazm.Normalizer()
text = normalizer.normalize(text)
```

```python
print(text)
```

```python
tokenizer = hazm.WordTokenizer()
tokens = tokenizer.tokenize(text)
```

```python
print(tokens)
```

```python
stop_words = hazm.stopwords_list()
```

```python
punc = list(string.punctuation + '؟،؛')
```

```python
stemmer = hazm.Stemmer()
```

```python
clean_txt = []
for token in tokens:
    if ( token not in stop_words) and (token not in punc):
        word = stemmer.stem(token)
        clean_txt.append(word)
```

```python
print(clean_txt)
```
