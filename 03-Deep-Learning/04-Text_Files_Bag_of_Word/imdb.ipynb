Now we are going to define a function to do all of this to gether

```python
def clean_text(text):
    #html to text format
    soup = BeautifulSoup(text)
    text = soup.getText()
    # to lowercase
    text = text.lower()
    # tokenize
    tokens = word_tokenize(text)
    # punctuatuions
    punc = list(string.punctuation)
    # stopwords and stemmer
    stop_words = stopwords.words('english')
    stemmer = PorterStemmer()
    filter_sent = []
    for token in tokens:
        if (token not in punc) and (token not in stop_words):
            word = stemmer.stem(token)
            filter_sent.append(word)
    # for having string with document format from filtered sentences
    text =' '.join(filter_sent)

    return text
```
