```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
sharpen_filter = np.array([
                       [0,   -0.5,   0],
                       [-0.5,  3, -0.5],
                       [0,   -0.5,   0]
                        ])

edge_filter = np.array([
                   [-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]
                      ])
```

```python
image = cv2.imread('cat1.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
image.shape
```

```python
image_sharpen = cv2.filter2D(image,-1,sharpen_filter)
plt.imshow(image_sharpen)
```

```python
image_edge = cv2.filter2D(image,-1,edge_filter)
plt.imshow(image_edge)
```
