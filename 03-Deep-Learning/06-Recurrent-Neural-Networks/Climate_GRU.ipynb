We want to pick dataset every 3 hours for seq. Then set the min on 0 and hour to numbers that dividable to 3.

```python
data = data[(data.index.minute == 0) & (data.index.hour % 3 == 0)]
```

```python
tempc = data['T (degC)']
var = data.drop(columns=['T (degC)','Tpot (K)'])
```

Scaling the dataset.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
var = scaler.fit_transform(var)
```

For 7 dayes and every day 8 seq we will have 56 seq.

Setting step=2 is going to jump one element in seq when want to start the next seq. In order seq one start from element 1 and end in 3. Seq 2 going to start form element 3 instate of element 2.

```python
seq_len = 56
step = 2
```

```python
sequences = []
targets = []

for i in range(0,len(data)-seq_len-7,step):
    seq = var[i:i+seq_len]
    target = tempc[i+seq_len+7]
    sequences.append(seq)
    targets.append(target)

sequences, targets = np.array(sequences), np.array(targets)
```

```python
sequences.shape
```

Because in time series we can not do the splitting random, then we do it manual with follow the order.

Train = 0.7 , Val = 0.2 , Test(final_test) = 0.1

```python
X_train, X_val, X_test = sequences[:8155], sequences[8155:10485], sequences[10485:]
y_train, y_val, y_test = targets[:8155], targets[8155:10485], targets[10485:]
```

```python
import tensorflow as tf
from tensorflow.keras.layers import GRU,Dense,Dropout
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(GRU(units=128,activation='tanh',recurrent_dropout=0.2,input_shape=(56,12),return_sequences=True))
model.add(GRU(units=128,activation='tanh',recurrent_dropout=0.2,dropout=0.3))
model.add(Dropout(0.5))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
opt = tf.optimizers.Adam(learning_rate=0.001)
loss = tf.losses.MeanAbsoluteError()
model.compile(optimizer=opt,loss=loss)
```

```python
model.fit(X_train,y_train,batch_size=50,epochs=40,validation_data=(X_val,y_val),verbose=2)
```

```python
model.evaluate(X_test,y_test)
```
