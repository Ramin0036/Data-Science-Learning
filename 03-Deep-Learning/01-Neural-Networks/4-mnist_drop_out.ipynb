```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import tensorflow as tf
from tensorflow import keras
```

```python
data = datasets.load_digits()
```

```python
X = data.data / 16
y = data.target
```

```python
y = keras.utils.to_categorical(y,num_classes=10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

An other way to prevent the overfit is Dropout. In this method we are going to disconnect some connections of units in other words deactiving some neurons.

```python
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(units=32 , activation='relu' , input_shape=(64,)))
model.add(keras.layers.Dropout(0.24))
model.add(keras.layers.Dense(units=32 , activation='relu'))
model.add(keras.layers.Dropout(0.24))
model.add(keras.layers.Dense(units=10 , activation='softmax'))
```

```python
model.summary()
```

```python
opt = tf.optimizers.Adam(learning_rate=0.01)
loss = tf.losses.CategoricalCrossentropy()

model.compile(optimizer=opt , loss= loss , metrics= [tf.metrics.CategoricalAccuracy()])
```

```python
# early_stopping
#early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss' , patience=5 ,restore_best_weights= True)
```

```python
model.fit(X_train,y_train,epochs=100 ,batch_size=50 , validation_split=0.2 , verbose=2)
```

We done it. mostly prevent the overfit with a simple dropout method.

```python
acc , loss = model.evaluate(X_test , y_test)
```
