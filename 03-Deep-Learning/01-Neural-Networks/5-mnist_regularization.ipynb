```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
import tensorflow as tf
from tensorflow import keras
```

```python
data = datasets.load_digits()
```

```python
from PIL import Image
```

```python
test_image = Image.open("D:\\python-venv\\NN\\test\\sample_image - Copy.png")
test_image.thumbnail((8,8))
test_image = np.asarray(test_image)
rgb_weight = np.array([0.3,0.4,0.3,0])
test_image = (test_image @ rgb_weight) / 255
test_im = test_image.reshape(1, 64)
```

```python
X = data.data / 16
y = data.target
```

```python
y = keras.utils.to_categorical(y,num_classes=10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

An other wa to prevent the overfit issue is using the regularizers. Its getting hard the training and learning the model and because of that its gonna reduce the speed of loss reduction.

```python
reg = keras.regularizers.l2(0.0001)
```

```python
model = tf.keras.models.Sequential()
model.add(keras.layers.Dense(units=64 , activation='relu' , input_shape=(64,),kernel_regularizer=reg))
model.add(keras.layers.Dropout(0.25))
model.add(keras.layers.Dense(units=64 , activation='relu',kernel_regularizer=reg))
model.add(keras.layers.Dropout(0.25))
model.add(keras.layers.Dense(units=10 , activation='softmax'))
```

```python
model.summary()
```

```python
opt = tf.optimizers.Adam(learning_rate=0.001)
loss = tf.losses.CategoricalCrossentropy()

model.compile(optimizer=opt , loss= loss , metrics= [tf.metrics.CategoricalAccuracy()])
```

```python
# early_stopping
#early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss' , patience=5 ,restore_best_weights= True)
```

```python
model.fit(X_train,y_train,epochs=100 ,batch_size=50 , validation_split=0.2 , verbose=2)
```

a good prevent from overfit.

```python
acc , loss = model.evaluate(X_test , y_test)
```
