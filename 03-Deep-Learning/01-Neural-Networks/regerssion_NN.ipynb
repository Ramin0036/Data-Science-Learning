```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense,Dropout,Flatten
```

```python
data = pd.read_csv("D:\\python-venv\\NN_resg\\Data\\concrete_data.csv")
```

```python
X = data.drop(['concrete_compressive_strength'] ,axis=1)
y = data['concrete_compressive_strength']
```

```python
y.shape
```

```python
X.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.3)
```

activation func and units changed than fashion_mnist because of that its regressoin and has just one output but in f_m was a classification problem with 10 classes.

```python
# model
model = tf.keras.models.Sequential()
model.add(Dense(units=256 , activation='relu' , input_shape = (8,)))
model.add(Dropout(0.3))
model.add(Dense(units=256 , activation='relu'))
model.add(Dense(units=1 , activation='linear'))
```

```python
# model_compiling
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = tf.keras.losses.MeanAbsoluteError()
model.compile(optimizer= opt , loss= loss)
```

```python
# model_fitting
model.fit(X_train , y_train , epochs= 200 , batch_size= 50 , verbose= 2 , validation_data=(X_test , y_test))
```

```python
model.evaluate(X,y)
```
