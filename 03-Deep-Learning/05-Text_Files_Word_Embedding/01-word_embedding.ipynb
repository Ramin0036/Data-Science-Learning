```python
documents = ['Keras is a deep learning API written in Python',
             'The core data structures of Keras are layers and models',
             'What you just saw is the most elementary way to use Keras',
             'If you need to, you can further configure your optimizer',
             'Computing the gradient of arbitrary differentiable expressions'
                ]
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(documents)
```

Thsts for inputdim in embedding layer in keras

```python
input_dim = len(tokenizer.word_index) + 1 
```

```python
tokenizer.word_counts
```

```python
encode = tokenizer.texts_to_sequences(documents)
```

```python
encode
```

```python
list_len = [] 
for seq in encode:
    list_len.append(len(seq))
```

```python
max_len = max(list_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

For making equal size for every one of sentences.
If you want to add those zero scores to the beginnig of each row you can just simply set the padding on pre

```python
padded_encode = pad_sequences(encode,maxlen=max_len,padding='post')
```

```python
padded_encode
```

```python
padded_encode.shape
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
```

oupput_dim is the num of feauters that we are givving the embedding to set for each word

```python
model = Sequential()
model.add(Embedding(input_dim=input_dim,output_dim=50))
```

```python
pred = model.predict(padded_encode)
```
