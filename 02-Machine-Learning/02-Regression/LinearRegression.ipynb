```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

Making a regression data with make_regression function.

```python
X,y = make_regression(n_samples=200,n_features=1,n_targets=1,noise=15,random_state=11)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=123)
```

```python
LR = LinearRegression()
LR.fit(X_train,y_train)
```

```python
LR.score(X_train,y_train)
```

```python
LR.score(X_test,y_test)
```

```python
from sklearn import metrics
```

```python
y_pred_test = LR.predict(X_test)
y_pred_train = LR.predict(X_train)
```

```python
metrics.root_mean_squared_error(y_test,y_pred_test)
```

```python
metrics.root_mean_squared_error(y_train,y_pred_train)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,y_pred_train,color="red")
plt.scatter(X_test,y_test)
plt.scatter(X_test,y_pred_test,color="black")
plt.show()
```
