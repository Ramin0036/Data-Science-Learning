```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X,y = make_regression(n_samples=50,n_features=1,n_targets=1,noise=20,random_state=5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=123)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
# KNN Regressor
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_reg = KNeighborsRegressor(n_neighbors=3)
knn_reg.fit(X_train,y_train)
```

```python
knn_reg.predict([[1]])
```

```python
knn_reg.score(X_train,y_train)
```

```python
knn_reg.score(X_test,y_test)
```

```python
# SVR
```

```python
from sklearn.svm import SVR
```

```python
svr_reg = SVR(kernel="linear",C=10,epsilon=50)
svr_reg.fit(X_train,y_train)
```

```python
svr_reg.score(X_train,y_train)
```

```python
svr_reg.score(X_test,y_test)
```

```python
y_pred = svr_reg.predict(X_train)
```

```python
plt.scatter(X_train,y_train)
plt.scatter(X_train,y_pred)

plt.show()
```

```python
# DTR
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
dtr_reg = DecisionTreeRegressor(max_depth=3)
dtr_reg.fit(X_train,y_train)
```

```python
y_pred = dtr_reg.predict(X_train)
```

```python
plt.scatter(X_train,y_train)
plt.scatter(X_train,y_pred)

plt.show()
```

```python
plt.figure(figsize=(35,20))
_ = plot_tree(dtr_reg,filled=True)
```
