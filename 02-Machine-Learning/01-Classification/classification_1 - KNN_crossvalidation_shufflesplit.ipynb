```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw = pd.read_csv("D:\\ML&DL\\heart.csv")
data = data_raw.copy()
```

```python
data.replace({"Age":0 , "RestingBP":0 , "Cholesterol":0 , "MaxHR":0 },np.nan , inplace=True)
```

```python
data.isna().sum()
```

```python
data_categorical = data.select_dtypes("object")
data_numerical = data.select_dtypes(["float64" , "int64"]).iloc[:,:-1]
y = data.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
label_enc = LabelEncoder()
```

```python
data_categorical = data_categorical.apply(label_enc.fit_transform)
```

```python
data_categorical.head()
```

```python
data_categorical.nunique()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe_enc = OneHotEncoder()
```

```python
data_categorical_bin = data_categorical[["Sex","ExerciseAngina"]]
```

```python
data_ohe = ohe_enc.fit_transform(data_categorical[["ChestPainType", "RestingECG" , "ST_Slope"]].values).toarray()
```

```python
data_ohe.shape
```

```python
data = np.concatenate([data_numerical.values , data_categorical_bin.values , data_ohe], axis = 1)
```

```python
pd.DataFrame(data)
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
data = imputer.fit_transform(data)
```

```python

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
X = data
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X,y)
```

```python
y_pred = knn.predict(X)
```

```python
print(f"y true :  {y[:10]}")
print(f"y pred :  {y_pred[:10]}")
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y,y_pred)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
```

```python
y_pred = knn.predict(X)
```

```python
accuracy_score(y,y_pred)
```

Calssification Evalution

Slice Method

```python
len(X[ : 600 , :])
len(y[ : 600])
y_train = y[: 600]
X_train = X[:600 , :]
y_test = y[600:]
X_test = X[600: , :]
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train , y_train)
```

```python
y_pred = knn.predict(X_train)
acc_pred = accuracy_score(y_train,y_pred)
print(acc_pred)
```

```python
y_pred = knn.predict(X_test)
acc_pred = accuracy_score(y_test , y_pred)
print(acc_pred)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=1234)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train , y_train)
```

```python
y_pred = knn.predict(X_train)
acc_pred = accuracy_score(y_train,y_pred)
print(acc_pred)
```

```python
y_pred = knn.predict(X_test)
acc_pred = accuracy_score(y_test , y_pred)
print(acc_pred)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_test , y_pred))
```

```python
metrics.confusion_matrix(y_test , y_pred)
```

Validation & Cross Validation

Using Validation for evaluate the classifier during the train proces.

```python
from sklearn.model_selection import cross_validate
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn_valid = cross_validate(knn , X , y , cv=4 , scoring="accuracy" , verbose=10)
```

Now we are going to use the shufflesplit because of that in cross validate we have data compression then the randomness broking the work.

```python
from sklearn.model_selection import ShuffleSplit
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
cv = ShuffleSplit(n_splits=5 , test_size=0.25 , random_state=1)
knn_valid = cross_validate(knn , X , y , cv=cv , scoring=["accuracy" , "f1"] , verbose=10 , return_train_score=True)
```
