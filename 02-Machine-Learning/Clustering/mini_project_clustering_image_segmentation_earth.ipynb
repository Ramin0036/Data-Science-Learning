```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_org = cv2.imread("D:\ML&DL\Clustering_project\image\data\pic\earth.jpg")
```

```python
plt.imshow(img_org)
plt.show()
```

```python
img = cv2.cvtColor(img_org,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
img_vectorized = img.reshape((-1,3))
```

```python
img_vectorized.shape
```

```python
from sklearn.cluster import KMeans
```

```python
segmentation = KMeans(n_clusters=4)
segmentation.fit(img_vectorized)
```

```python
np.unique(segmentation.labels_)
```

```python
labels = segmentation.labels_.reshape(img.shape[:2])
```

```python
labels.shape
```

```python
plt.figure(figsize=(15,10))
plt.imshow(labels,cmap='gray')
plt.show()
```

```python
# label 1
labels[labels!=2] = 0
```

```python
plt.figure(figsize=(15,10))
plt.imshow(labels,cmap='gray')
plt.show()
```
