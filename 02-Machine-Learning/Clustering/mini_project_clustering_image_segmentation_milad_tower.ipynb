```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
org_image = cv2.imread("D:\ML&DL\Clustering_project\image\data\pic\Milad-Tower.jpg")
```

```python
plt.imshow(org_image)
plt.show()
```

```python
RGB_image = cv2.cvtColor(org_image , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_image)
plt.show()
```

```python
RGB_image.shape
```

```python
data = RGB_image.reshape((-1,3))
```

```python
x1 = data.T[0]
x2 = data.T[1]
x3 = data.T[2]
```

```python
data.shape
```

```python
from sklearn.cluster import DBSCAN , KMeans
```

```python
model = KMeans(n_clusters=2)
model.fit(data)
```

```python
model.labels_.shape
```

```python
labels = model.labels_.reshape(RGB_image.shape[:2])
```

```python
plt.imshow(labels , cmap= 'gray')
plt.show()
```

```python
c1 = model.cluster_centers_[0]
c2 = model.cluster_centers_[1]
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.scatter(c1[0] , c1[1] ,c1[2] , color = 'red' , s=2000 , marker= 'x')
ax.scatter(c2[0] , c2[1] ,c2[2] , color = 'red' , s=2000 , marker= 'x')
ax.scatter(x1 , x2 , x3)
plt.show()
```

```python
target = cv2.findNonZero(cv2.inRange(labels , 1 , 1))
```

```python
target.shape
```

```python
target = cv2.findNonZero(cv2.inRange(labels , 1,1)).reshape((-1,2))
```

```python
target.shape
```

```python
RGB_image_new = RGB_image.copy()
```

```python
for i in target:
    RGB_image_new[i[1] , i[0] -200] = RGB_image_new[i[1] , i[0]] 
```

```python
plt.imshow(RGB_image_new)
plt.show()
```
