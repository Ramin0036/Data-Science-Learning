```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Customers.csv")
```

```python
data.rename(columns={
    "Annual Income (k$)": "Annual Income",
    "Spending Score (1-100)": "Spending Score"
            },
           inplace=True)
```

```python
data.head()
```

```python
data.Age.min()
```

```python
data.Age.max()
```

```python
age_bins = [0,20,30,40,50,60,70]
```

```python
data["Age Segment"] = pd.cut(x= data.Age,bins=age_bins)
```

```python
data.head()
```

```python
data.groupby("Age Segment").boxplot(figsize=(15,15))
plt.show()
```

```python
data2 = data.drop(columns=["CustomerID","Age"]).copy()
```

```python
data2.groupby("Age Segment").boxplot(figsize=(15,15))
plt.show()
```

```python
grp_age_gender = data2.groupby(["Age Segment","Gender"])
```

```python
for group , data in grp_age_gender:
    data.boxplot()
    title = "Age :" + str(group[0]) + "--Gender :" + str(group[1])
    plt.title(title)
    plt.show()
```
