```python
def MAD_outlier_detector(data , columns= None , threshold=3):
    if columns == None:
        data_analyzer = data.drop(columns=['class'])
    else:
        data_analyzer = data[columns]

    median = np.median(data_analyzer , axis = 0)
    MAD = np.median(np.abs(data_analyzer - median) , axis=0)
    mi = (0.6745 * (data_analyzer - median)) / MAD

    y_predict_index = set(np.where(np.abs(mi)>threshold)[0])
    y_true_index = set(np.where(data['class'] == 0)[0])

    true_detection = y_predict_index.intersection(y_true_index)
    hads_dorost = len(true_detection) / len(y_true_index)
    hads_nadorost = (len(y_predict_index) - len(true_detection)) / (len(data) - len(y_true_index))

    report = {"ture_guess" : hads_dorost , "false_guess":hads_nadorost}
    return report
```

```python
MAD_outlier_detector(data , columns=['same_srv_rate' , 'dst_host_srv_count' , 'flag' , 'same_srv_rate'] , threshold= 2.8)
```
