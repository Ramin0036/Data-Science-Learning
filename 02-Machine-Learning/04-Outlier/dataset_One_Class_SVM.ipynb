```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("D:\ML&DL\outlier\data\data.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
```

```python
data.head()
```

```python
data['class'][data['class'] == 0] = -1
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
data.iloc[:,:-1] = scaler.fit_transform(data.iloc[:,:-1])
```

```python
data.head()
```

```python
data_inlier = data[data['class'] == 1]
data_outlier = data[data['class'] == -1]
```

```python
data_test = pd.concat([data_inlier[:1500] , data_outlier[:1500]])
data_train = data_inlier[1500: ]
```

```python
X_train = data_train.drop(columns=['class'])
y_train = data_train['class']
```

```python
X_test = data_test.drop(columns=['class'])
y_test = data_test['class']
```

```python
from sklearn.svm import OneClassSVM
```

```python
model = OneClassSVM(nu= 0.06)
model.fit(X_train)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred_train = model.predict(X_train)
```

```python
np.unique(y_pred_train , return_counts=True)
```

```python
print(classification_report(y_train , y_pred_train))
```

```python
y_pred_test = model.predict(X_test)
```

```python
np.unique(y_pred_test , return_counts=True)
```

```python
print(classification_report(y_test , y_pred_test))
```
