According to the explanations in the readme section, using the boxplot method.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =np.array([
    [2,4,6],
    [3,5,7],
    [10,8,-2],
    [17,2.5,8],
    [1.5,-8,-9],
    [12,51,50],
    [2.5,4,-23]
    ])
```

```python
np.percentile(data , 50)
```

```python
np.median(data)
```

```python
Q1 = np.percentile(data , 25)
Q3 = np.percentile(data , 75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
```

```python
data > upper_bound
```

```python
data < lower_bound
```

```python
(data > upper_bound ) | (data < lower_bound)
```

```python
plt.boxplot(data)
plt.show()
```

```python
sns.boxplot(data)
plt.show()
```
