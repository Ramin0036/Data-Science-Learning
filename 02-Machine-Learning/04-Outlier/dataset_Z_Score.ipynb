```python
def zscore_outlier_detector(data , columns= None , threshold=3):
    if columns == None:
        data_analyzer = data.drop(columns=['class'])
    else:
        data_analyzer = data[columns]

    z_scores =zscore(data_analyzer , axis=0)
    y_hads_index = set(np.where(np.abs(z_scores)>threshold)[0])
    y_true_index = set(np.where(data['class'] == 0)[0])

    true_detection = y_hads_index.intersection(y_true_index)
    hads_dorost = len(true_detection) / len(y_true_index)
    hads_nadorost = (len(y_hads_index) - len(true_detection)) / (len(data) - len(y_true_index))

    report = {"true_guess" : hads_dorost , "false_guess":hads_nadorost}
    return report
```

```python
zscore_outlier_detector(data , threshold=1.27)
```
