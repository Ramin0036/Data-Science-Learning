```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
```

Making some points with make_regression method.

```python
X , y= make_regression(n_samples=10 , n_features=1 , noise=5)
```

```python
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
y_predict = model.predict(X)
```

Visualizing the regressor.

```python
plt.scatter(X,y)
plt.plot(X , y_predict)
plt.show()
```

Adding 2 outlier point to dataset to see the impact of outlier data.

```python
X_new = np.concatenate((X,[[1.7],[0]]))
y_new = np.concatenate((y,[-100,200]))
```

```python
plt.scatter(X_new , y_new)
plt.show()
```

again making regression.

```python
model_1 = LinearRegression()
model_1.fit(X_new , y_new)
```

```python
y_predict_new = model_1.predict(X_new)
```

Now we can see the outlier impact in regression func.

```python
plt.scatter(X_new , y_new)
plt.plot(X , y_predict)
plt.plot(X_new , y_predict_new)
plt.show()
```
