Assume this dataset manual made.

```python
data =np.array([
    [2,4,6],
    [3,5,7],
    [10,8,-2],
    [17,2.5,8],
    [1.5,-8,-9],
    [12,51,50],
    [2.5,4,-23]
    ])
```

```python
from scipy.stats import zscore
```

```python
z_scores = zscore(data , axis=0)
```

If condition in below was true for one of arrayes in any row, then this row is concidered outlier fully. 

```python
(z_scores > 2) | (z_scores<-2)
```
