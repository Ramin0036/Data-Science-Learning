```python
import matplotlib.pylab as plt
from sklearn import datasets
import numpy as np
```

Making circle data.

```python
datasets.make_circles(n_samples=500 , noise=0.02 , factor=0.3)
```

```python
data , label = datasets.make_circles(n_samples=500 , noise=0.02 , factor=0.3)
```

```python
plt.scatter(data.T[0] , data.T[1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
model.fit(data)
```

```python
cluster0 = data[model.labels_ == 0]
cluster1 = data[model.labels_ == 1]
```

```python
plt.scatter(cluster0.T[0] , cluster0.T[1])
plt.scatter(cluster1.T[0],cluster1.T[1])
plt.show()
```

Kmeans can not handle this type of datasets. This is the weakness of KMeans.
