```python
import matplotlib.pylab as plt
from sklearn import datasets
import numpy as np
```

```python
data , label = datasets.make_circles(n_samples=500 , noise=0.02 , factor=0.3)
```

```python
plt.scatter(data.T[0] , data.T[1])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
model.fit(data)
```

```python
cluster0 = data[model.labels_ == 0]
cluster1 = data[model.labels_ == 1]
```

```python
plt.scatter(cluster0.T[0] , cluster0.T[1])
plt.scatter(cluster1.T[0],cluster1.T[1])
plt.show()
```

```python
data1 , label1 = datasets.make_moons(n_samples=500 , noise=0.02 )
```

```python
plt.scatter(data1.T[0] , data1.T[1])
plt.show()
```

```python
model1 = KMeans(n_clusters=2)
model1.fit(data1)
```

```python
cluster0_1 = data1[model1.labels_ == 0]
cluster1_1 = data1[model1.labels_ == 1]
```

```python
plt.scatter(cluster0_1.T[0] , cluster0_1.T[1])
plt.scatter(cluster1_1.T[0],cluster1_1.T[1])
plt.show()
```

Its clear that KMeans can not clustering this type of datasets then we are going to use DBScan.

```python
from sklearn.cluster import DBSCAN
```

```python
DB = DBSCAN()
DB.fit(data)
```

```python
cluster0DB = data[DB.labels_ == 0]
cluster1DB = data[DB.labels_ == 1]
```

```python
plt.scatter(cluster0DB.T[0] , cluster0DB.T[1])
plt.scatter(cluster1DB.T[0],cluster1DB.T[1])
plt.show()
```

```python
DB1 = DBSCAN(eps=0.1)
DB1.fit(data1)
```

```python
cluster0DB_1 = data1[DB1.labels_ == 0]
cluster1DB_1 = data1[DB1.labels_ == 1]
```

```python
plt.scatter(cluster0DB_1.T[0] , cluster0DB_1.T[1])
plt.scatter(cluster1DB_1.T[0],cluster1DB_1.T[1])
plt.show()
```
