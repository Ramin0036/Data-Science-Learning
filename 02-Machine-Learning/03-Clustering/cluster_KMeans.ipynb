```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Creating random points.

```python
g1 = np.random.normal(7,2,size=(100 , 2))
g2 = np.random.normal(15,2,size=(100 , 2))
data = np.concatenate((g1 ,g2))
np.random.shuffle(data)
```

```python
plt.scatter(data.T[0] , data.T[1])
```

one of clustering algorithms is KMeans.

```python
from sklearn.cluster import KMeans
```

We have 2 cluster then we have to set cluster number on 2.

```python
model = KMeans(n_clusters=2)
model.fit(data)
```

Getting the clusters centers.

```python
centroid = model.cluster_centers_
```

```python
plt.scatter(data.T[0] , data.T[1])
plt.scatter(centroid[0,0] , centroid[0,1] , marker= 'X' , s= 150 , color = "black")
plt.scatter(centroid[1,0] , centroid[1,1] , marker= 'X' , s= 150 , color = "black")
plt.scatter(data.T[0].mean() , data.T[1].mean() , marker= '*' , s=150 , color = "red")
```

Labaling the clusters.

```python
cluster0 = data[model.labels_ == 0]
cluster1 = data[model.labels_ == 1]
```
