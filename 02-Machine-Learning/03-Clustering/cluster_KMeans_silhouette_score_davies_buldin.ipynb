```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
g1 = np.random.normal(7,1.5,size=(100 , 2))
g2 = np.random.normal(15,1.5,size=(100 , 2))
g3_x = np.random.normal(7,1.5,size=(120 , 1))
g3_y = np.random.normal(15,1.5,size=(120 , 1))
g3 = np.concatenate((g3_x , g3_y) , axis=1)
data = np.concatenate((g1 ,g2 , g3))
np.random.shuffle(data)
```

```python
plt.scatter(data.T[0] , data.T[1])
```

```python
from sklearn.cluster import KMeans
```

```python
#wcss >>>>>> model.inertia_
```

Using the method in below we are going to find the best cluster number for algorithm.

```python
wcss = []
for i in range(1,15):
    model = KMeans(n_clusters=i)
    model.fit(data)
    wcss.append(model.inertia_)
```

```python
plt.plot(range(1,15),wcss)
plt.show()
```

Elbow showing the best n_cluster is 3. 

```python
model_behine = KMeans(n_clusters=3)
model_behine.fit(data)
```

Now we are going to evaluate or scoring the models, using silhouette_score and davies_bouldin_score.

```python
from sklearn.metrics import silhouette_score
```

```python
sil_score = []
for i in range(2,15):
    model = KMeans(n_clusters=i)
    model.fit(data)
    sil_score.append(silhouette_score(data , model.labels_))
```

```python
plt.plot(range(2,15) , sil_score)
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
davis_buldin_score = []
for i in range(2,15):
    model = KMeans(n_clusters=i)
    model.fit(data)
    davis_buldin_score.append(davies_bouldin_score(data , model.labels_))
```

```python
plt.plot(range(2,15) , davis_buldin_score)
plt.show()
```
