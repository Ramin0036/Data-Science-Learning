```python
import numpy as np
from sklearn.decomposition import FastICA
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X, y = make_circles(n_samples=1000,random_state=42,noise=0.02)
```

```python
y
```

```python
X
```

# ICA

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
fica = FastICA(n_components=1, random_state=123)
fica_result = fica.fit_transform(X)
```

```python
plt.scatter(fica_result, [1 for _ in range(0, 1000)], c=y)
plt.show()
```

```python
plt.scatter(fica_result, range(0, 1000), c=y)
plt.show()
```
