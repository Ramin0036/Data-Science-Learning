```python
import numpy as np
from sklearn.manifold import Isomap
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
```

```python
X, y = make_moons(n_samples=1000,random_state=42,noise=0.03)
```

```python
y
```

```python
X
```

# Isomap

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
isomap = Isomap(n_components=1,n_neighbors=10)
fica_result = isomap.fit_transform(X)
```

```python
plt.scatter(fica_result, [1 for _ in range(0, 1000)], c=y)
plt.show()
```

```python
plt.scatter(fica_result, range(0, 1000), c=y)
plt.show()
```
