```python
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
X, y = make_circles(n_samples=1000,random_state=42,noise=0.02)
```

```python
y
```

```python
X
```

# PCA

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
pca = PCA(n_components=1, random_state=123)
pca_result = pca.fit_transform(X)
```

```python
plt.scatter(pca_result, [1 for _ in range(0, 1000)], c=y)
plt.show()
```

```python
plt.scatter(pca_result, range(0, 1000), c=y)
plt.show()
```

# Kernel PCA

```python
kernel_pca = KernelPCA(n_components=1,kernel='rbf' ,random_state=123)
kernel_pca_result = kernel_pca.fit_transform(X)
```

```python
plt.scatter(kernel_pca_result, range(0, 1000), c=y)
plt.show()
```
