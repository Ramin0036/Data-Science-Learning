```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing dataset

```python
data_raw  = pd.read_csv("heart.csv")
data = data_raw.copy()
```

For the general information of datatset

```python
data.head()
```

Below syntax is for checking that the Cholesterol column has any 0 values?

```python
data[data.Cholesterol == 0]
```

Result of pervious syntax showing that 172 smapels (rows) has 0 values 

Below syntax is gonig to change 0 values of Cholesterol column to nan(null)

```python
data.replace({"Cholesterol":0},np.nan,inplace=True)
```

```python
data.isna().sum()
```

Imputer is a function to change nan values with mean of non_zero neihgbors for solving the nun values issue

```python
from sklearn.impute import SimpleImputer, KNNImputer
```

```python
simple_imp = SimpleImputer(missing_values= np.nan , strategy="mean")
```

```python
simple_imp.fit(data[["Cholesterol"]])
```

```python
data["Cholesterol"] = simple_imp.transform(data[["Cholesterol"]])
```

After applying SimpleImputer there is no missing_values or nun values in data set

```python
data.isna().sum()
```
